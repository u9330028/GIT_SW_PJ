```
import torchvision.datasets as dsets
import torchvision.transforms as transforms

mnist_train = dsets.MNIST(root='MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
mnist_test = dsets.MNIST(root='MNIST_data/',
                         train=False,
                         transform=transforms.ToTensor(),
                         download=True)
```

```
import torch
USE_CUDA = torch.cuda.is_available()
DEVICE = torch.device("cuda" if USE_CUDA else "cpu")
DEVICE
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = mnist_train.data.view(-1,28*28).float().to(DEVICE)
y_train = mnist_train.targets.to(DEVICE)
x_test = mnist_test.data.view(-1, 28*28).float().to(DEVICE)
y_test = mnist_test.targets.to(DEVICE)

model = nn.Linear(28*28, 10).to(DEVICE)

optimizer = torch.optim.Adam(model.parameters())

for epoch in range(201):
    z = model(x_train)
    cost = F.cross_entropy(z, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
  
    with torch.no_grad():
        test_z = model(x_test)
        test_cost = F.cross_entropy(test_z, y_test)

    if epoch % 10 == 0:
        with torch.no_grad():
            z_test = model(x_test)
            accuracy = 100*(y_test==z_test.argmax(1)).sum().item()/len(z_test)
            cost_test = F.cross_entropy(z_test, y_test)
            print("epoch: {}, cost: {:.6f}, test: {:.6f}, acc: {:.1f}%".format(epoch, cost.item(), cost_test.item(),accuracy))
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = mnist_train.data.view(-1,28*28).float().to(DEVICE)
y_train = mnist_train.targets.to(DEVICE)
x_test = mnist_test.data.view(-1, 28*28).float().to(DEVICE)
y_test = mnist_test.targets.to(DEVICE)

layer1 = nn.Linear(28*28, 256).to(DEVICE)
bn1= nn.BatchNorm1d(256).to(DEVICE)
layer2 = nn.Linear(256, 256).to(DEVICE)
bn2= nn.BatchNorm1d(256).to(DEVICE)
layer3 = nn.Linear(256, 10).to(DEVICE)
relu = nn.ReLU()

nn.init.xavier_normal_(layer1.weight, 0.1)
nn.init.xavier_normal_(layer2.weight, 0.1)
nn.init.xavier_normal_(layer3.weight, 0.1)

model = nn.Sequential(layer1, bn1, relu, layer2, bn2, relu, layer3)

optimizer = torch.optim.Adam(model.parameters())

for epoch in range(201):
    z = model(x_train)
    cost = F.cross_entropy(z, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 10 == 0:
        with torch.no_grad():
            z_test = model(x_test)
            accuracy = 100*(y_test==z_test.argmax(1)).sum().item()/len(z_test)
            cost_test = F.cross_entropy(z_test, y_test)
            print("epoch: {}, cost: {:.6f}, test: {:.6f}, acc: {:.1f}%".format(epoch, cost.item(), cost_test.item(),accuracy))
```
